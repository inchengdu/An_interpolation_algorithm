Dataset production

```python
import numpy as np
```

```python
lens=200   # data set length
```

```python
data_full=np.load("")    # Preprocessed data storage address
```

```python
print(data_full.shape)
print(data_full)
```

```python
def lack_math(raw,lack_c,clear_c):
    full_data=raw.copy()
    full_data[clear_c]=0
    full_data=np.delete(full_data,lack_c,axis=0)
    real=raw[lack_c]
    zero=np.zeros((1,lens))
    full_data=np.append(full_data,zero,axis=0)
    return full_data,real
```

```python
data_5=data_full[52,:,:]
full_data,real=lack_math(data_5,[5],[5,6])
```

```python
print(full_data.shape)
print(full_data[:,0:2])
```

```python
print(real.shape)
```

```python
np.save('',real)                # Label save address
np.save('',full_data)           # Data storage address with bad sectors
np.savetxt('',full_data)        # The data with bad sectors is saved in txt format, which is convenient for importing into EEGLAB
```
